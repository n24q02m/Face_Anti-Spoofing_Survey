```python
import os
import cv2
import matplotlib.pyplot as plt

# Đường dẫn đến thư mục ImposterRaw
base_path = "/kaggle/input/nuaaaa/raw/ImposterRaw"

# Lấy thư mục con đầu tiên
subfolder = os.path.join(base_path, os.listdir(base_path)[0])  # Thư mục con đầu tiên
image_files = os.listdir(subfolder)[:5]  # Lấy 5 tệp đầu tiên

# Hiển thị 5 ảnh đầu tiên
plt.figure(figsize=(15, 5))  # Kích thước cửa sổ hiển thị
for i, image_file in enumerate(image_files):
    img_path = os.path.join(subfolder, image_file)
    img = cv2.imread(img_path)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)  # Chuyển sang định dạng RGB để hiển thị
    
    plt.subplot(1, 5, i + 1)  # Tạo subplot cho mỗi ảnh
    plt.imshow(img)
    plt.axis('off')  # Tắt trục tọa độ
    plt.title(f"Image {i + 1}")

plt.show()
```

```python
import os
import matplotlib.pyplot as plt

# Đường dẫn đến các thư mục chính
client_path = "/kaggle/input/nuaaaa/raw/ClientRaw"
imposter_path = "/kaggle/input/nuaaaa/raw/ImposterRaw"

# Hàm đếm tổng số ảnh trong một thư mục (bao gồm thư mục con)
def count_images(folder_path):
    total_images = 0
    for subfolder in os.listdir(folder_path):
        subfolder_path = os.path.join(folder_path, subfolder)
        if os.path.isdir(subfolder_path):  # Kiểm tra thư mục con
            total_images += len(os.listdir(subfolder_path))  # Đếm số tệp trong thư mục con
    return total_images

# Đếm số lượng ảnh trong ClientRaw và ImposterRaw
client_image_count = count_images(client_path)
imposter_image_count = count_images(imposter_path)

# Vẽ biểu đồ
labels = ['ClientRaw', 'ImposterRaw']
counts = [client_image_count, imposter_image_count]

plt.bar(labels, counts, color=['blue', 'orange'])
plt.xlabel('Thư mục')
plt.ylabel('Số lượng ảnh')
plt.title('So sánh số lượng ảnh giữa ClientRaw và ImposterRaw')
plt.ylim(0, max(counts) + 100)  # Đặt giới hạn trục y
for i, count in enumerate(counts):
    plt.text(i, count + 5, str(count), ha='center', fontsize=10)

plt.show()
```

# Kích thước ảnh

```python
import cv2

def analyze_image_sizes(folder_path):
    sizes = []
    for subfolder in os.listdir(folder_path):
        subfolder_path = os.path.join(folder_path, subfolder)
        if os.path.isdir(subfolder_path):
            for img_name in os.listdir(subfolder_path):
                img_path = os.path.join(subfolder_path, img_name)
                img = cv2.imread(img_path)
                if img is not None:
                    sizes.append(img.shape[:2])  # Lấy chiều cao và chiều rộng
    return sizes

client_sizes = analyze_image_sizes(client_path)
imposter_sizes = analyze_image_sizes(imposter_path)

# Đếm phân phối kích thước ảnh
from collections import Counter
print("Kích thước ảnh ClientRaw:", Counter(client_sizes).most_common(5))
print("Kích thước ảnh ImposterRaw:", Counter(imposter_sizes).most_common(5))
```

# Phân tích màu sắc ảnh

```python
import numpy as np

def calculate_histogram(folder_path):
    histograms = []
    for subfolder in os.listdir(folder_path):
        subfolder_path = os.path.join(folder_path, subfolder)
        if os.path.isdir(subfolder_path):
            for img_name in os.listdir(subfolder_path)[:10]:  # Lấy 10 ảnh đầu tiên từ mỗi thư mục con
                img_path = os.path.join(subfolder_path, img_name)
                img = cv2.imread(img_path)
                if img is not None:
                    hist = cv2.calcHist([img], [0, 1, 2], None, [256, 256, 256], [0, 256, 0, 256, 0, 256])
                    histograms.append(hist)
    return np.mean(histograms, axis=0)

client_histogram = calculate_histogram(client_path)
imposter_histogram = calculate_histogram(imposter_path)

# So sánh histogram giữa Client và Imposter
plt.plot(client_histogram.ravel(), label='ClientRaw', color='blue')
plt.plot(imposter_histogram.ravel(), label='ImposterRaw', color='orange')
plt.xlabel('Pixel Intensity')
plt.ylabel('Frequency')
plt.title('Histogram màu')
plt.legend()
plt.show()
```

Tập trung ở hai đầu: Biểu đồ cho thấy phần lớn pixel nằm ở cường độ rất thấp (gần 0) và rất cao (giá trị tối đa). Điều này có thể xảy ra nếu ảnh chứa nhiều vùng tối và sáng rõ rệt, ví dụ như ảnh đen trắng hoặc có độ tương phản cao.
Tương đồng giữa hai tập dữ liệu: Đường màu xanh (ClientRaw) và cam (ImposterRaw) gần như chồng lên nhau, chứng tỏ phân phối màu sắc của hai tập dữ liệu này tương đối giống nhau.
Cao điểm ở đầu và cuối: Đỉnh cao ở các giá trị pixel nhỏ cho thấy các ảnh có thể chứa nhiều vùng tối hoặc bóng mờ.
=> hai tập dữ liệu có đặc điểm về màu sắc hoặc ánh sáng tương đồng, phù hợp với giả định hai tập có cùng bối cảnh chụp hoặc điều kiện ánh sáng.

```python
import hashlib

def calculate_hash(img_path):
    with open(img_path, 'rb') as f:
        return hashlib.md5(f.read()).hexdigest()

def find_duplicates(folder_path):
    hashes = {}
    duplicates = []
    for subfolder in os.listdir(folder_path):
        subfolder_path = os.path.join(folder_path, subfolder)
        if os.path.isdir(subfolder_path):
            for img_name in os.listdir(subfolder_path):
                img_path = os.path.join(subfolder_path, img_name)
                img_hash = calculate_hash(img_path)
                if img_hash in hashes:
                    duplicates.append(img_path)
                else:
                    hashes[img_hash] = img_path
    return duplicates

duplicates_client = find_duplicates(client_path)
duplicates_imposter = find_duplicates(imposter_path)

print(f"Số ảnh trùng lặp trong ClientRaw: {len(duplicates_client)}")
print(f"Số ảnh trùng lặp trong ImposterRaw: {len(duplicates_imposter)}")
```
